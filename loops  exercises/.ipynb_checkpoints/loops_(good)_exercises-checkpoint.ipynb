```python
%%timeit -n 1000 #number of loops -r 5 #number of runs #run timeit multiple times to check for high variance
print([x for x in range(10) if x%2==1])
```

List comprehension is faster to run than 'for' loops but it has a higher variance in mean run time.

```python
initial=100
current=initial
years=0
while current>0.5*initial:
    current*=0.95
    years+=1
print(f'It will take {years} years.')
```

```python
co2_data = [
    (2000, 369.55), (2001, 371.14), (2002, 373.28), 
    (2003, 375.80), (2004, 377.52), (2005, 379.80), 
    (2006, 381.90), (2007, 383.79), (2008, 385.60), 
    (2009, 387.43), (2010, 389.90), (2011, 391.65), 
    (2012, 393.85), (2013, 396.52), (2014, 398.65),
    (2015, 400.83), (2016, 404.24), (2017, 406.55), 
    (2018, 408.52), (2019, 411.44), (2020, 414.24)
]

for x in range(1,len(co2_data)):
    year, level=co2_data[x]
    prev_year, prev_level=co2_data[x-1]
    change=level-prev_level
    if change>=3:
        print(f'Year:{year}, change:{change:.2f}')
```
